# Seaborn is a Data Visualization Library for Python
           ## It is build on top of the Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
pengu = sns.load_dataset("penguins")
pengu
```

## Seaborn_Scatter plot

```python
sns.set_theme()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="sex")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size", hue="sex")
```

## Seaborn_Line plot

```python
flights = sns.load_dataset("flights")
flights
```

```python
# Automatically takes the 'mean' value if not specified.
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
sns.lineplot(data=flights, x="year", y="passengers", estimator="sum")
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

```python
trips = sns.load_dataset("taxis", parse_dates=["pickup", "dropoff"])
trips
```

```python
trips["hour"] = trips["pickup"].dt.hour
trips
```

```python
sns.lineplot(data=trips, x="hour", y="total", hue="payment")
```

```python
sns.lineplot(data=trips, x="hour", y="total", hue="payment", style="color", ci=None)
```

## Seaborn_relplots

```python
sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="sex")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="sex", hue="smoker")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="sex", row="time", hue="smoker")
```

```python
sns.relplot(data=trips, x="hour", y="total", kind="line", col="pickup_borough", hue="payment")
```

```python
sns.relplot(
    data=trips, 
    x="hour", 
    y="total", 
    kind="line", 
    col="pickup_borough", 
    hue="payment",
    row="dropoff_borough"
)
```

## Changing size

```python
# plt.figsize argument still effects the X, Y,(relational) plot showed below
plt.figure(figsize=(8,5))
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

```python
# plt.figsize argument (is still valid) will no effect on 'relplots'
sns.relplot(
    data=tips, 
    x="total_bill", 
    y="tip", 
    col="day", 
    hue="smoker",
    row="sex",
    height=2,
    aspect=1.5
)
```

## Distributions_ Histplot

```python
sns.histplot(
    data=tips, 
    x="tip",
    hue="time"
)
```

```python
sns.histplot(
    data=tips, 
    x="tip",
    hue="smoker",
    multiple="stack"
)
```

```python
sns.histplot(
    data=tips, 
    x="tip",
    hue="smoker",
    multiple="dodge"
)
```

```python
sns.histplot(data = pengu, x="body_mass_g", binwidth=300)
```

```python
sns.histplot(data = pengu, x="body_mass_g", bins=20)
```

```python
sns.histplot(data = pengu, 
             x="body_mass_g", 
             bins=20, 
             hue="species",
             multiple="stack"
            )
```

```python
sns.histplot(data = pengu, 
             x="body_mass_g", 
             bins=20, 
             hue="species",
             multiple="stack",
             element="step"
            )
```

```python
sns.histplot(data=tips, x="tip", kde=True)
```

```python
sns.histplot(data = pengu, 
             x="body_mass_g", 
             bins=20, 
             hue="species",
             multiple="stack",
             kde=True
            )
```

## KDE_plot (Kernal density estimation)

```python
sns.kdeplot(data=pengu, x="body_mass_g")
```

```python
sns.kdeplot(data=pengu, x="body_mass_g", hue="species")
```

```python
sns.kdeplot(
    data=pengu, 
    x="body_mass_g", 
    hue="species",
    bw_adjust=0.5
)
```

```python
sns.kdeplot(
    data=pengu, 
    x="body_mass_g", 
    hue="species",
    multiple="stack"
)
```

## Rugplot

```python
sns.rugplot(data=tips, 
            x="tip",
            height=0.3
           )
```

```python
sns.rugplot(data=tips, 
            y="tip",
            height=0.5
           )
```

```python
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill", height=0.07)
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip", height=0.03)
```

```python
diamonds = sns.load_dataset("diamonds")
sns.scatterplot(data=diamonds, x="carat", y="price", s=5)
sns.rugplot(data=diamonds, x="carat", y="price", lw=1, alpha=.005)
```

## displot (Distribution)_ fihgure level plotting method

```python
sns.displot(kind="hist", data=pengu, 
            x="body_mass_g", height=3,
            aspect=2
           )
```

```python
sns.displot(kind="hist", 
            data=pengu, 
            x="body_mass_g", 
            hue="sex",
            col="species",
            row="island",
            height=3,
           )
```

```python
sns.displot(data=tips, kind="kde", 
            x="tip", col="time", rug=True)
```

```python
sns.displot(data=tips, kind="kde", 
            x="total_bill", y="tip", rug=True)
```
