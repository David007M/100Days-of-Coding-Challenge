```python
import numpy as np
nums = np.arange(5)
nums
```

```python
plt.figure()
plt.plot(nums,nums)
plt.figure()
plt.plot(nums,nums*nums)
plt.figure()
plt.plot(nums,nums**3)
```

```python
salaries = [55000,65000,72000,90000,115000,150000]
ages = [20,25,30,32,40,45]
plt.plot(ages,salaries, color="blue", linewidth=5, marker="p", markersize="15", markerfacecolor="yellow")
plt.title("Company Salaries")
plt.figure()
plt.plot(nums,nums, linestyle="dashed", color="red")
plt.title("Line Stuff", loc="right")
```

```python
salaries = [55000,65000,72000,90000,115000,150000]
ages = [20,25,30,32,40,45]
plt.plot(ages,salaries, color="blue", linewidth=5, marker="p", markersize="15", markerfacecolor="yellow")
plt.title("Company Salaries")
plt.xlabel("Employee Age", labelpad=30)
plt.ylabel("Average Salary", labelpad=30, size=20)
```

```python
plt.figure(figsize=(8,8))
salaries = [55000,65000,72000,90000,115000,150000]
ages = [20,25,30,32,40,45]
plt.plot(ages,salaries, color="blue", marker="p", markersize="10", markerfacecolor="red")
plt.title("Company Salaries")
plt.xlabel("Employee Age", labelpad=20)
plt.ylabel("Average Salary", labelpad=20)
plt.xticks([20,25,30,35,40,45,50])
plt.yticks([50000,80000,100000,140000], labels=["50k", "80k", "100k", "140k"], size=20)
```

```python
plt.figure(figsize=(8,8))
salaries = [55000,65000,72000,90000,115000,150000]
ages = [20,25,30,32,40,45]
plt.plot(ages,salaries, color="blue", marker="p", markersize="10", markerfacecolor="red")
plt.xlim(25,45)
plt.ylim(65000,150000)
```
