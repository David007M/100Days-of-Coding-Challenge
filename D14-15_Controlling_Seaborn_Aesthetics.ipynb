```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## The Figure-Level catplot( ) Method

```python
titanic = sns.load_dataset("titanic")
trips = sns.load_dataset("taxis")
```

```python
sns.set_style()
```

```python
sns.catplot(data=titanic, x="sex", y="survived", kind="bar")
```

```python
sns.catplot(data=titanic, x="sex", y="survived", kind="bar", col="pclass")
plt.show()
```

```python
sns.catplot(data=titanic, x="pclass", y="survived", kind="bar", hue="sex")
plt.show()
```

```python
sns.catplot(
    data=trips, 
    kind="strip", 
    x="pickup_borough", 
    y="distance", 
    col="color",
    aspect=1.5
)
plt.show()
```

```python
sns.catplot(data=trips, kind="violin", x="pickup_borough", y="distance", hue="color", split=True)
plt.show()
```

```python
sns.catplot(data=titanic, 
           kind="bar",
           x="who",
           y="survived",
           col="pclass",
           ci=None,
           row="embarked"
           )
plt.show()
```

# Cutomizing Seaborn Aesthetics

```python
# Main themes in Seaborn
# "white", "dark", "whitegrid", "darkgrid", "ticks"
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.set_style("white")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.set_style("dark")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.set_style("ticks")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.axes_style()
```

```python
sns.axes_style("white")
```

```python
sns.set_style({'axes.facecolor': '#FC427B', 'grid.color': '#9AECDB'})
```

```python
sns.set_style("ticks", {'axes.facecolor': '#FC427B', 'grid.color': '#9AECDB'})
```

```python
# Getting back to original theme
sns.set_style("darkgrid")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
# Spines
sns.set_style("white")
```

```python
sns.histplot(data=tips, x="total_bill")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill")
sns.despine() # will remove the 'top' and 'right' spines
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill")
sns.despine(bottom=True) # will remove the 'top' and 'right' + 'bottom' spines
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill")
sns.despine(bottom=True, top=False) 
plt.show()
```
